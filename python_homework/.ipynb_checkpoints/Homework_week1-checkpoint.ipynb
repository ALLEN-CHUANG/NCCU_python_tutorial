```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
time = [2, 4, 1, 3]
item = [1, 2, 3, 4]
```

# 溫馨小提示

### title: 給你的plot一個專屬的名字吧。
### xticks: 用來取代X軸的項目名。
### label: 給Ｘ軸一個自己的名字。

```python
plt.title("My plan in Freshman year")
#plt.xkcd() 
plt.xlabel('Item')
plt.ylabel('Hours')
plt.xticks([1, 2, 3, 4], [r'PT', r'SW', r'LOVE', r'CLUB'])
plt.plot(item, time, marker='o', ms=10)
```

# 『加分題』溫馨小提示  ^.^

## 利用pie這個函數畫出圓餅圖吧～
### explode:可以調整分離出來的部分與圓餅圖的距離。
### color:可以調整個圓餅區塊的顏色。
### label:幫個區塊自己加上名字吧。

```python
explode = (0.2, 0, 0, 0) 
color = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue'] 
label = ['PART-TIME JOB','SCHOOLWORK','LOVE','CLUB']
plt.pie(time,explode=explode, labels=label, colors=color,
        autopct='%1.1f%%', shadow=True, startangle=140)
```
